```python
import pandas as pd
df= pd.read_csv('restaurant_customer_satisfaction.csv')
df.columns
```

```python
print(df.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.boxplot(x='HighSatisfaction', y='Age', data=df)
plt.title("Müşteri Yaşı ve Memnuniyet İlişkisi")
plt.xlabel("Memnuniyet")
plt.ylabel("Yaş")
plt.show()
```

```python
sns.histplot(df['Income'], bins=30, kde=True)  # Sütun adı büyük harfle yazılmalı
plt.title("Gelir Dağılımı")
plt.xlabel("Gelir")
plt.ylabel("Frekans")
plt.show()
```

```python
sns.boxplot(x='HighSatisfaction', y='ServiceRating', data=df)
plt.title("Hizmet Puanı ve Memnuniyet İlişkisi")
plt.xlabel("Memnuniyet (0: Düşük, 1: Yüksek)")
plt.ylabel("Hizmet Puanı")
plt.show()
```

```python
sns.histplot(df['AverageSpend'], bins=20, kde=True)
plt.title("Ortalama Harcama Dağılımı")
plt.xlabel("Ortalama Harcama")
plt.ylabel("Frekans")
plt.show()
```

```python
sns.countplot(x='MealType', hue='HighSatisfaction', data=df)
plt.title("Yemek Türü ve Memnuniyet İlişkisi")
plt.xlabel("Yemek Türü")
plt.ylabel("Frekans")
plt.legend(title="Memnuniyet (0: Düşük, 1: Yüksek)")
plt.show()
```

```python
sns.countplot(x='Gender', hue='HighSatisfaction', data=df)
plt.title("Cinsiyet ve Memnuniyet İlişkisi")
plt.xlabel("Cinsiyet")
plt.ylabel("Frekans")
plt.legend(title="Memnuniyet (0: Düşük, 1: Yüksek)")
plt.show()
```

```python
sns.scatterplot(x='GroupSize', y='AverageSpend', data=df)
plt.title("Grup Boyutu ve Harcama Miktarı İlişkisi")
plt.xlabel("Grup Boyutu")
plt.ylabel("Ortalama Harcama")
plt.show()
```
